```python
import spacy
```

```python
nlp = spacy.load('en')

# create list to track tags for each review
tag_list = list()
tag_idx_list = list()

# create dictionaries to track vacabulary indexes
voc_tag = dict()
voc_tag_cnt = 0

for review in df.review.tolist():
    comm_tag_list = list()
    comm_tag_idx_list = list()
    doc = nlp(review)
    for token in doc:
        w = token.text
        if str.strip(w) not in ['','.']:      
            # tag int (fine pos tag)
            g = token.tag_
            if voc_tag.get(g,voc_tag_cnt) == voc_tag_cnt:
                voc_tag[g] = voc_tag_cnt 
                voc_tag_cnt += 1
            comm_tag_list.append(g)           
            comm_tag_idx_list.append(voc_tag[g]) 
    tag_list.append(comm_tag_list)
    tag_idx_list.append(comm_tag_idx_list)

```

```python
X_tag = np.array(tag_list)
X_tag_idx = np.array(tag_idx_list)
```

```python
voc_size_tag = len(voc_tag)
```

## CNN model

Build the CNN model

```python
def conv_model(output_dim=64, filter_size=128, window_size=3, stride=1, pool_size=2, dense_dim=16):
    
    # for word
    krn_size = output_dim * window_size
    input1 = Input(shape=(max_len,), dtype='int32')
    embed1 = Embedding(voc_size+1, output_dim, input_length=max_len, embeddings_initializer='random_uniform')(input1)
    C1 = Conv1D(filter_size, kernel_size=krn_size ,padding='same', strides=(stride), activation='relu')(embed1)
    C1 = Dropout(0.5)(C1)
    M1 = MaxPooling1D(pool_size=(pool_size), padding='same')(C1)
    F1 = Flatten()(M1)
    D1 = Dense(dense_dim, activation='relu')(F1)
    D1 = Dropout(0.5)(D1)
    
    # for POS tag
    input2 = Input(shape=(max_len,), dtype='int32')
    embed2 = Embedding(voc_size_tag+1, int(output_dim/2), input_length=max_len, \
                       embeddings_initializer='random_uniform')(input2)
    C2 = Conv1D(int(filter_size/2), kernel_size=int(krn_size/2) ,padding='same', \
                       strides=(stride), activation='relu')(embed2)
    C2 = Dropout(0.5)(C2)
    M2 = MaxPooling1D(pool_size=(pool_size), padding='same')(C2)
    F2 = Flatten()(M2)
    D2 = Dense(1, activation='relu')(F2)
    D2 = Dropout(0.5)(D2)
    
    # Merge branches by concatenating along axis 1
    L1 = concatenate([D1, D2], axis=1)
    G1= Dense(dense_dim, activation='relu')(L1)
    G1 = Dropout(0.5)(G1)

    pred = Dense(1, kernel_regularizer=regularizers.l2(0.01),\
                 activity_regularizer=regularizers.l1(0.01),\
                 activation='sigmoid')(G1)
    
    model = Model(inputs=[input1, input2], outputs=pred)
    
    #sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    return model
    
conv_model().summary()
```

```python
EPOCHS = 10
BATCH_SIZE = 64
fold=8

# concatenates
X = np.hstack((sequences, tag_seqs))
y = np.array(df.lable.tolist())
```

K-Fold for validation

```python
kfold = StratifiedKFold(n_splits=fold, shuffle=True, random_state=7)
```

## Model performance : prediction accuracy

```python
acr = list()
i = 0
for train_index, valid_index in kfold.split(X, y):
    i += 1
    X_train, X_valid = X[train_index], X[valid_index]
    y_train, y_valid = y[train_index], y[valid_index]
    model = conv_model()
    # divide input back into sequences and tag_seqs
    model.fit([X_train[:,:max_len], X_train[:,max_len:]], y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=2, \
              validation_data=([X_valid[:,:max_len], X_valid[:,max_len:]], y_valid),\
              callbacks=[EarlyStopping(patience=1, monitor='val_loss')])
    y_h = model.predict([X_valid[:,:max_len], X_valid[:,max_len:]])
    y_pred = [round(pred) for pred in y_h.reshape(-1)]
    
    acr.append(accuracy_score(y_valid, y_pred))
```

```python
print('Average accuracy ', np.mean(np.array(acr)))
```
